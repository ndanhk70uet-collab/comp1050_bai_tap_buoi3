```python
#W5A1
a, b= map(int, input().split())
def Max_Of_Two(a,b):
    if a>b:
        return a
    else:
        return b
print(Max_Of_Two(a,b))
```

```python
#W5A2
def swap(a,b):
    a,b= b,a
    return a,b
a, b= map(int, input().split())
m, n= swap(a,b)
print(m,n)
```

```python
#W5A3
def isprime(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i==0:
            return False
    return True
n=int(input())
print(isprime(n))
```

```python
#W5A4
def perfect_number(n):
    count=0
    for i in range (1,n):
        if n%i==0:
            count=count+i
    if count==n:
        return True
    return False
n=int(input())
print(perfect_number(n))
```

```python
#W5A5
def find_first(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i+1
    return -1
a=list(map(int, input().split()))
n=int(input())
print(find_first(a, n))
```

```python
#W5A6
def factorial(n):
    fac=1
    if n>0:
        for i in range (1, n+1):
            fac=fac*i
        return fac
n=int(input())
print(factorial(n))
```

```python
#W5A7
def cal(a, b, op):
    if op== '+':
     return a+b
    elif op== '-':
       return a-b
    elif op== '*':
       return a*b
    elif op== '/':
       return a/b
a, op, b=input().split()
a=float(a)
b=float(b)
print(f"{cal(a, b, op):.2f}")
```

```python
#W5A8
def hamming(a, b):
    a=bin(a^b)
    b=list(str(a))
    count=0
    for i in b:
        if i=='1':
            count+=1
    return count
a,b =map(int, input().split())
print(hamming(a,b))
```

```python
#W5A9
def sum_digit(n):
    s=0
    while n>0:
        s=s+n%10
        n=n//10
    return s
a=int(input())
print(sum_digit(a))
```

```python
#W5A10
def is_isomorphic(a, b):
    map1 = {}
    map2 = {}

    for x, y in zip(a, b):
        if x in map1 and map1[x] != y:
            return False
        if y in map2 and map2[y] != x:
            return False
        map1[x] = y
        map2[y] = x

    return True

a, b = input().split()
print(is_isomorphic(a, b))
```
