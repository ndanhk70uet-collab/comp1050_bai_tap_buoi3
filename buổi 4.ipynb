```python
#W4A8
def reverse(n):
    s=0
    while n>0:
        s=s*10+n%10
        n=n//10
    return s
def đối_xứng(a):
    a=str(a)
    for i in range(0,len(a)):
        if a[i]!=a[len(a)-1-i]:
            return False
    return True
b=int(input())
count=0
while đối_xứng(b)==False:
    b=b+reverse(b)
    count+=1
print(count, b)
```

```python
#W4A10
def collatz(n):
    count=1
    while n!=1:
        if n%2==0:
            n/=2
            count+=1
        else:
            n=3*n+1
            count+=1
    return count
n=int(input())
max=0
best=0
for i in range(1,n+1):
    if collatz(i)>max:
        max=collatz(i)
        best=i
print(best, max)
```

```python
#W4A14
a, b=map(int, input().split())
def GCD(a,b):
    if a>b:
        for i in range(b,0,-1):
            if a%i==0 and b%i==0:
                break
        return i
    if b>a:
        for i in range(a,0,-1):
            if a%i==0 and b%i==0:
                break
        return i
print(GCD(a,b))
```
