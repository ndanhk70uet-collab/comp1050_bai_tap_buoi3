```python
#W6A1
a = list(map (int , input().split()))
ans = []
seen = {}
for i in range(len(a)):
    if not a[i] in seen:
        seen[a[i]] = 0
for i in range(len(a)):
    if not seen[a[i]]:
        ans.append(a[i])
        seen[a[i]] = 1
print(ans)
```

```python
#W6A9
def add(dict0 , a):
    key = ''
    value = ''
    for i in range(len(dict0)):
        if dict0[i].isalnum() and key =='':
            while i < len(dict0) and dict0[i].isalnum():
                key += dict0[i]
                i += 1
        elif dict0[i].isalnum() and key !='':
            while i < len(dict0) and dict0[i].isalnum():
                value += dict0[i]
                i += 1
        elif dict0[i] == ',' or dict0[i] == '}':
            if key != '' and value != '':
                if key not in a:
                    a[key] = int(value)
                else:
                    a[key] += int(value)
                key = ''
                value = ''
dict1 = input()
dict2 = input()
a = {}
add(dict1 , a)
add(dict2 , a)
sorted_keys = sorted(a.keys())
print(a)
```

```python
#W6A10
a = list(map(int , input().split()))
k = int(input())
dict = {}
ans = []
for i in a:
    if ( (k- i) in dict ) and  ( (k - i , i) not in ans ):
            ans.append((k - i , i))
    if i not in dict:
        dict[i] = 1
ans.sort()
print(ans)
```

```python
#W6A12
a = list(map(int , input().split()))
map = {}
for i in a:
    if i not in map:
        map[i] = 1
    else:
        map[i] += 1
ans = a[0] 
for key , value in map.items():
    if value > map[ans]:
        ans = key
    elif value == map[ans] and key < ans:
        ans = key 
print(ans)
```
